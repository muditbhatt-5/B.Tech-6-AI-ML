## Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

## Step 2. Import the dataset

```python
data = pd.read_csv('iphones.csv')

print(data.head())
```

## Step 3. Import LinearRegression model from linear_model family 

```python
from sklearn.linear_model import LinearRegression
```

## Step 4. Create LinearRegression Model

```python
model = LinearRegression()
```

## Step 5. Splitting dataset in to input and output

```python
x = data[['iphonenumber']]  
y = data['price']    

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

## Step 6. Fit the Model (Training)

```python
model.fit(X_train, y_train)

print("Coefficient:", model.coef_)
print("Intercept:", model.intercept_)
```

## Step 7. Prdict the value

```python
y_pred = model.predict(X_test)

for actual, predicted in zip(y_test[:5], y_pred[:5]):
    print("Actual Value => ", actual)
    print("Predicted Value => ", predicted)
```

## Step 8. Result Visualization

```python
plt.scatter(X_test, y_test, color='blue', label='Actual')
plt.plot(X_test, y_pred, color='red', label='Predicted')
plt.title('Actual vs Predicted')
plt.xlabel('X_column')
plt.ylabel('y_column')
plt.legend()
plt.show()

print("Mean Squared Error =>", mean_squared_error(y_test, y_pred))
print("R² Score => ", r2_score(y_test, y_pred))
```

## Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

## Step 2.  Read the Data

```python
data = pd.read_csv('Salary_Data.csv')

print(data.head())
```

## Step 3. View Data using Scatter plot 

```python
plt.scatter(data['YearsExperience'], data['Salary'], color='blue')
plt.title('Salary X Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

## Step 4. Splitting dataset in to input and output

```python
input = data[['YearsExperience']]  
output = data['Salary']             

X_train, X_test, y_train, y_test = train_test_split(input, output, test_size=0.2, random_state=42)
```

## Step 5. Splitting dataset in to input and output

```python
input = data[['YearsExperience']]  
output = data['Salary']             

X_train, X_test, y_train, y_test = train_test_split(input, output, test_size=0.2, random_state=42)
```

## Step 6. Create LinearRegression Model

```python
model = LinearRegression()
```

## Step 7. Fit the Model (Training)

```python
model.fit(X_train, y_train)

print("Coefficient: ", model.coef_)
print("Intercept: ", model.intercept_)
```

##  Step 6. Predicting the Test set results

```python
y_pred = model.predict(X_test)

for actual, predicted in zip(y_test[:5], y_pred[:5]):
    for i in range(0, 5):
        print("Test Case #",i)
        print("Actual: ", actual)
        print("Predicted: ", predicted)
```

##  Step 7. Visualising the Training set results

```python
plt.scatter(X_train, y_train, color='blue', label='Actual')
plt.plot(X_train, model.predict(X_train), color='red', label='Regression Line')
plt.title('Training Set: Salary vs. Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

##  Step 8. Visualising the Test set results

```python
plt.scatter(X_test, y_test, color='green', label='Actual')
plt.plot(X_test, y_pred, color='red', label='Regression Line')
plt.title('Test Set: Salary vs. Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()

print("Mean Squared Error: ", mean_squared_error(y_test, y_pred))
print("R² Score: ", r2_score(y_test, y_pred))
```
